```python
import tensorflow as tf
import pickle
from tensorflow.keras.models import load_model

model = load_model('churn_model.h5')
scaler = pickle.load(open('standard_scaler.pkl', 'rb'))
label_encoder_gender = pickle.load(open('label_encoder.pkl', 'rb'))
one_hot_encoder_geo = pickle.load(open('one_hot_encoder.pkl', 'rb'))
```

```python
input_data={
    'CreditScore': 0,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
import pandas as pd
categorical_df=pd.DataFrame(one_hot_encoder_geo.transform([[input_data['Geography']]]).toarray(), columns=one_hot_encoder_geo.get_feature_names_out(['Geography']))
```

```python
input_df= pd.concat([pd.DataFrame([input_data]).reset_index(drop=True).drop(['Geography'], axis=1), categorical_df], axis=1)
```

```python
input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
```

```python
input_df
```

```python
scaled_input = scaler.transform(input_df)
scaled_input
```

```python
y_pred_probability=model.predict(scaled_input)[0][0]
```

```python
type(y_pred_probability)
```

```python
y_pred= (y_pred_probability > 0.5).astype(int)
```
